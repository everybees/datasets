Quiz 1: 


In examining Africa and Asia, for example, is there a positive correlation between the percentage of men in active employment as a proportion of the total men in the country's population and that country's GDP?


Consider information from the World Bank, and the following countries as samples: Nigeria, Ghana, Morocco, Botswana, Kenya, India, China, Afghanistan, Lebanon and Cambodia for the period 1999 to 2019.


From your analysis of the dataset, which of these countries has the strongest positive correlation and what is the pearson coefficient?


A. Nigeria, 0.9903583503162041

B. Ghana, 0.8310467199786551

C. Lebanon, 0.9452918830894971

D. India, 0.98851215310157

E. Morocco, 0.969748897475768


Expected response: D (India, 0.98851215310157)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import pearsonr
```

Employment data

```python
employment_data = pd.read_csv('https://raw.githubusercontent.com/everybees/datasets/main/employment.csv')
```

```python
pd.set_option('display.max_columns',None)
employment_data.drop(['Indicator Name','Unnamed: 66', 'Country Code', 'Indicator Code', '2020', '2021'], axis=1, inplace=True)
employment_data
```

```python
employment_range = [0] + [i for i in range(40,61)]
employment_info = employment_data.iloc[:,employment_range]
employment_info
```

```python
employment = pd.melt(employment_info, id_vars=['Country Name'], var_name='year', value_name='active_male_employment(in_percentage)')
employment
```

```python
sns.barplot(data=employment, y='active_male_employment(in_percentage)', x='year')
```

GDP data

```python
gdp_data = pd.read_csv('https://raw.githubusercontent.com/everybees/datasets/main/gdp.csv')
gdp_data
```

```python
gdp_range = [0] + [i for i in range(43,64)]
gdp_info = gdp_data.iloc[:,gdp_range]
gdp_info
```

```python
gdp = pd.melt(gdp_info, id_vars=['Country Name'], var_name='year', value_name='gdp(in_US_dollars)')
gdp
```

Total male population data

```python
male_pop = pd.read_csv('https://raw.githubusercontent.com/everybees/datasets/main/male_population.csv')
male_pop
```

```python
male_pop_range = [0] + [i for i in range(43,64)]
male_pop_info = male_pop.iloc[:,male_pop_range]
male_pop_info
```

```python
male_population = pd.melt(male_pop_info, id_vars=['Country Name'], var_name='year', value_name='total_male_population(in_thousands)')
male_population
```

```python
employ_gdp = pd.merge(employment, gdp, on=['Country Name', 'year'])
```

```python
employ_gdp_pop = pd.merge(employ_gdp, male_population, on=['Country Name', 'year'])
employ_gdp_pop.head()
```

```python
employ_gdp_pop['active_male(in_thousands)'] = (employ_gdp_pop['active_male_employment(in_percentage)']/100) * employ_gdp_pop['total_male_population(in_thousands)']
employ_gdp_pop.head()
```

```python
employ_gdp_population = employ_gdp_pop.reindex(columns=['Country Name','year','active_male_employment(in_percentage)','total_male_population(in_thousands)','active_male(in_thousands)','gdp(in_US_dollars)'])
employ_gdp_population.head()
```

```python
employ_gdp_population.drop(['active_male_employment(in_percentage)'],axis=1,inplace=True)
```

```python
employ_gdp_population.head()
```

African countries sample

```python
nigeria = employ_gdp_pop.loc[employ_gdp_pop['Country Name'] == 'Nigeria']
ghana = employ_gdp_population.loc[employ_gdp_pop['Country Name'] == 'Ghana']
morocco = employ_gdp_pop.loc[employ_gdp['Country Name'] == 'Morocco']
botswana = employ_gdp_population.loc[employ_gdp_pop['Country Name'] == 'Botswana']
kenya = employ_gdp_population.loc[employ_gdp['Country Name'] == 'Kenya']
```

```python
A = nigeria['active_male(in_thousands)'].corr(nigeria['gdp(in_US_dollars)'])
B = ghana['active_male(in_thousands)'].corr(ghana['gdp(in_US_dollars)'])
C = morocco['active_male(in_thousands)'].corr(morocco['gdp(in_US_dollars)'])
D = botswana['active_male(in_thousands)'].corr(botswana['gdp(in_US_dollars)'])
E = kenya['active_male(in_thousands)'].corr(kenya['gdp(in_US_dollars)'])
print(A,B,C,D,E)
options = [A,B,C,D,E]
max_country = max(options)
print(f'The country with the most positive correlation between active male employment and gdp is {max_country}')
```

Asian countries sample

```python
india = employ_gdp_population.loc[employ_gdp['Country Name'] == 'India']
china = employ_gdp_population.loc[employ_gdp['Country Name'] == 'China']
afghanistan = employ_gdp_population.loc[employ_gdp['Country Name'] == 'Afghanistan']
lebanon = employ_gdp_population.loc[employ_gdp['Country Name'] == 'Lebanon']
cambodia = employ_gdp_population.loc[employ_gdp['Country Name'] == 'Cambodia']
```

```python
india_cor = india['active_male(in_thousands)'].corr(india['gdp(in_US_dollars)'])
china_cor = china['active_male(in_thousands)'].corr(china['gdp(in_US_dollars)'])
afghanistan_cor = afghanistan['active_male(in_thousands)'].corr(afghanistan['gdp(in_US_dollars)'])
lebanon_cor = lebanon['active_male(in_thousands)'].corr(lebanon['gdp(in_US_dollars)'])
cambodia_cor = cambodia['active_male(in_thousands)'].corr(cambodia['gdp(in_US_dollars)'])
print(india_cor,china_cor,afghanistan_cor,lebanon_cor,cambodia_cor)
options = [india_cor,china_cor,afghanistan_cor,lebanon_cor,cambodia_cor]
max_country = max(options)
print(f'The country with the most positive correlation between active male employment and gdp is {max_country}')
```
